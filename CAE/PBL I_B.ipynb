```
import math #LucasKaminski desenvolvido no google colabs

pi = math.pi #declaração do pi
K = 60000 #definindo constante
x = (2/3) #power do rh
n = 0.011 #coeficiente pvc


D = float(input("Insira o diâmetro do canal semicircular, em milímetros: "))
i = float(input("Insira a declividade presente no canal, em milímetros: "))

di_inicial = D * 0.005
di_final = D * 0.02 

if di_inicial<=i<=di_final:
  im = (i / 1000) #milimetros em metros
  imraiz = math.sqrt(im)

  Dm = (D / 1000) 

  S = ((pi*math.pow(Dm,2))/8) #area molhada em m²

  Rh = (Dm / 4) #raio hidraulico em m
  Rhpow = math.pow(Rh,x)

  Q = ((K*S*Rhpow*imraiz)/n)
  Q2 = round(Q, 2)
  print(Q2)
else:
  print("Valor da declividade incorreto.")
```
