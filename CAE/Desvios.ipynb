```
import math

print("Bem vindo ao cálculo de raízes de uma equação quadrada")
a = float(input(print("Insira o coeficiente a:")))
b = float(input(print("Insira o coeficiente b:")))
c = float(input(print("Insira o coeficiente c:")))

delta = (math.pow(b,2) - 4 * a * c)
if delta < 0:
  print("Não existe raízes reais para os coeficientes informados")
elif delta == 0:
  x1 = (-b / (2*a))
  print("Existe somente uma raiz para os coeficientes, sendo ela o",x1)
else:
  x1 = ((-b + math.sqrt(delta)) / (2*a))
  x2 = ((-b - math.sqrt(delta)) / (2*a))
  print("Existem duas raizes para os coeficientes, sendo o",x1,"e o",x2)
```
