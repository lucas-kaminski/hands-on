```
import math #exercicios

X = [] #definindo listas
Y = []
pi = math.pi

n = int(input(print("Insira o número de vezes a ser calculado")))

for i in range(0,n): # for 
  x = float(input(print("Insira o", i+1,+"º valor:")))
  X.append(x)

X.sort(reverse=False)

for i in range(0,n):
  y = X[i]
  gy = math.log2(y)+((math.pow(pi,3))/(2*math.exp(y)))
  Y.append(gy)

Y.sort(reverse=False)

for i in range(len(Y)):
    print("x:",round(X[i],4),"y:",round(Y[i],4))
```
