```
import math #pbl 1A
x = float(input(print("Digite a distância percorrida: ")))

if x > 0: 
  fx = (2 * math.sin(x) + 30) 
  Esquerdo = fx
  print("A temperatura do pneu esquerdo é", Esquerdo)
  
  gx = (4 * math.cos(x) + 30)
  Direito = gx
  print("A temperatura do pneu direito é", Direito)

  Diferenca = (Esquerdo - Direito)
  print(Diferenca > 0)

else:
  print("Insira uma distância positiva")
```
