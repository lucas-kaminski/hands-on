```
import math

print("Valor de Pi: " + str(math.pi))
angulo = int(input("Angulo em graus: "))

x = math.radians(angulo) #converte para radianos
print("Angulo em radianos: " + str(x))

y = math.sin(x)
print("Seno do angulo: " + str(y))

w = math.cos(x) 
print("Seno do angulo: " + str(w))

z = math.tan(x) 
print("Tangente do angulo: " + str(z))
```
